```python
KEN_RED = [248, 0, 0]
# KEN_GREEN = [88, 176, 40]
KEN_GREEN = [72, 192,  24]

ARM = [16,16,8]
```

```python

# 检测图像中与给定颜色相近的像素，并返回第一个匹配的像素位置
color = np.array(KEN_GREEN)
frame = test[80:200,:]
# Detect the Green color of Ken

# axis=2，计算沿着第三个轴（即颜色通道轴）的向量范数。将每个像素的 RGB 颜色值视为一个向量，并计算每个像素的颜色与给定颜色值之间的欧氏距离
diff = np.linalg.norm(np.array(frame) - np.array(color), axis=2)
# print("diff", diff)
mask = diff < 90
# Return the index where the red color is detected
coordinates = mask.nonzero()
# print("coordinates", coordinates)
# Add back the vertical offset
first_match = (coordinates[1][0], coordinates[0][0] + 80)
print("first_match", first_match)
plt.imshow(mask)
```

```python
color = np.array(KEN_RED)
frame = test[80:200,:]
# Detect the red color of Ken

diff = np.linalg.norm(np.array(frame) - np.array(color), axis=2)
# print("diff", diff)
mask = diff < 90
# Return the index where the red color is detected
coordinates = mask.nonzero()
# print("coordinates", coordinates)
# Add back the vertical offset
first_match = (coordinates[1][0], coordinates[0][0] + 80)
print("first_match", first_match)

plt.imshow(mask)
```

```python
color_red = np.array(KEN_RED)
color_green = np.array(KEN_GREEN)
color_arm = np.array(ARM)

diff_red = np.linalg.norm(np.array(frame) - np.array(color_red) , axis=2)
mask_red = diff_red < 90

diff_green = np.linalg.norm(np.array(frame) - np.array(color_green) , axis=2)
mask_green = diff_green < 90

plt.imshow(mask_red+mask_green)

```
